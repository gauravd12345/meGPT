## Creating Vocabulary

```python
folder = 'sherlock'
words = set()
for filename in os.listdir(folder):
    filepath = os.path.join(folder, filename)

    if os.path.isfile(filepath):
        with open(filepath, 'r', encoding='utf-8') as f:
            text = f.read()
            words.update(text)


tokens = sorted(words)
print(''.join(tokens))
```
